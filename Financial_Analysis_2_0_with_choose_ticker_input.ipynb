**Time series portfolio analysis**

```
df = ticktock.history(period="1d",start="2016-03-26",end="2022-04-26")
```

```
df
```

```
df.plot(figsize =(20,20),subplots=True)
```

```
df.head()
```

```
closedata = pd.DataFrame(df['Close'])
```

```
closedata.plot(figsize=(12,12),subplots=True)
```

```
percent_change = closedata['DIS']/(closedata['DIS'].shift(1)-1)
```

```
percent_change.plot(figsize = (12,12))
```

```
percent_change_nflx = closedata['NFLX']/(closedata['NFLX'].shift(1)-1)
```

```
percent_change_nflx.plot(figsize=(12,12))
```

```
percent_change_t = closedata['T']/(closedata['T'].shift(1)-1)
```

```
percent_change_t.plot(figsize=(12,12))
```

```
percent_change_nflx.plot.density(figsize=(12,12))
```

```
percent_change.plot.density(figsize=(12,12))
```

```
percent_change_t.plot.density(figsize=(12,12))
```
